```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
images=np.load('data/images.npy')
labels=np.load('data/labels.npy')
```

```python
# print (class_zero.shape)
class_zero=images[labels==0]
plt.imshow(class_zero[0],cmap='gray')
```

```python
image_list={}
images_show=[]
labels_list=[0,1,2,3,4]
for label in labels_list:
    image_list[label]=images[labels==label]
    plt.figure()
    plt.title("Current Label-"+str(label))
    plt.imshow(image_list[label][0],cmap='gray')
```
